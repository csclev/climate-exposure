## [The Impact of Tornadoes and Damaging Winds on County‑Level Income Growth in the United States](https://link.springer.com/article/10.1007/s13753-024-00605-2)

NOAA Severe Weather Database Files
This study examined the impact of tornadoes and damaging winds on county-level income growth in the United States of America between 1969 and 2023

**Regression:**

Change in Personal Income / Income Per Capita = 
+ Dummy variables for Tornado at time T
+ Dummy variables for winds at time T
+ Log Population density
+ number of events at that period
+ Time fixed effects
+ County fixed effects
+ State fixed effects
